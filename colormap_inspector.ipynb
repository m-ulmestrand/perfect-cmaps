### Import necessary functions

```python
from perfect_cmaps.colormaps import get_cmap, plot_images_with_colormap
from perfect_cmaps.color_utils import plot_colormap
from perfect_cmaps.storage import get_test_img_path
```

### Define some test images
These are by default pre-existing example images, but you can try your colormap out on any image. 

```python
test_images = [
    "charcoal4_49721153188_o-scaled.jpg",
    "Lily-Pollen.png",
    "melamine-foam_49722002717_o-scaled.jpg",
    "PhenomPharos-Gallery_Pyrite.jpg",
    "PhenomXLGallery_5.jpg",
]

test_image_path = get_test_img_path() / "sem"
test_images = [test_image_path / img_name for img_name in test_images]
```

### Get desired colormap and define number of points etc.
The number of points in the colormap can be set for desired granularity. The smoothing applies some amount of gaussian filtering to the colormap, beteen 0 and 1. Some more settings are available for generating colormaps, see the function `get_colormap()`.

```python
cmap_name = "viridish"
num_points = 1000
smoothing = 0.03
```

### Generate colormap

```python
cmap = get_cmap(cmap_name, num_points, smoothing=smoothing)
```

### Plot colormap gradient and test colormap on eample images

```python
plot_colormap(cmap, num_points)
plot_images_with_colormap(test_images, cmap)
```
